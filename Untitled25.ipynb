##Bar Plot

```python
x=df['Degree %']
y=df['Package(LPA)']
```

```python
plt.bar(x,y,color='skyblue', width=0.5, align='center', edgecolor='gray')
plt.title("Bar plot for company with their respective package")
plt.ylim(0,15)
plt.show()
```

```python
p=df["Gender"]
q=df["status"]
```

```python
plt.barh(p,q, height=0.5,color='skyblue')
plt.title("status of male and female students")
plt.xlabel("status")
plt.ylabel("gender")
plt.show()
```

Pie Chart


```python
plt.figure(figsize=(13,13))
a=df['name of company'].value_counts()
labels = df['name of company'].unique()
plt.pie(a,labels=labels,autopct=lambda p: '{:.0f}'.format(p * sum(a) / 100))
plt.title("Pie chart for company names")
plt.show()
```

```python
plt.figure(figsize=(10,10))
b=df["Degree stream"].value_counts()
label=df["Degree stream"].unique()
plt.pie(b,labels=label,autopct=lambda p: '{:.0f}'.format(p * sum(b) / 100))
plt.title("Pie chart for Degree stream of students")
plt.show()
```

HISTOGRAM


```python
r=df["SKILLS"]
s=df["Work exp"]
```

```python
plt.hist(r)
plt.title("SKILLS")
plt.xlabel("Skill set")
plt.ylabel("range")
```

```python
plt.hist(s)
plt.title("Work experience")
plt.xlabel("experience")
plt.ylabel("range")
```

LINE PLOT

```python
X=df['specialisation']
plt.plot(X, y, marker='o', linestyle='-', color='b', label='Data Points')
plt.xlabel("specialization")
plt.ylabel("expected package")
plt.legend()
plt.show()
```

BOX PLOT

```python
Y=df["name of company"].value_counts()
plt.boxplot(Y)
plt.title("box plot for name of company ")
plt.show()
```

```python
Z=df["Package(LPA)"]
plt.boxplot(Z)
plt.title("package")
plt.ylabel("LPA")
plt.show()
```

```python
data1 = df["name of company"].value_counts()
data2 = df["CGPA"].value_counts()
plt.boxplot([data1, data2])
plt.title("outliers analysis for name of company vs package")
plt.xlabel("company")
plt.ylabel("CGPA")
plt.show()
```
